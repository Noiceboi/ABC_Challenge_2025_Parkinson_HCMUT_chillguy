```python
acc = pd.read_csv('acc_sorted_by_time.csv')
```

```python
test = pd.read_csv('TestActivities_UTC0.csv')
```

```python
test['Started'] = pd.to_datetime(test['Started'])
```

```python
test['Finished'] = pd.to_datetime(test['Finished'])
test['Updated'] = pd.to_datetime(test['Updated'])
```

```python
test_time = test['Updated'].drop_duplicates().reset_index(drop=True)

# Display the first few values of the updated variable
test_time
```

```python
acc['Timestamp'] = pd.to_datetime(acc['Timestamp'])
```

```python
acc
```

```python
acc['Timestamp_Just_Minute'] = acc['Timestamp'].apply(lambda x: x.replace(second=0))
```

```python
acc['Timestamp_Just_Minute'] = acc['Timestamp_Just_Minute'].apply(lambda x: x.replace(microsecond=0))
```

```python
acc
```

```python
# Iterate over each 'Updated' time in test_time
for updated_time in test_time:
    # Filter rows in acc where 'Timestamp' matches the current 'Updated' time
    matched_rows = acc[acc['Timestamp_Just_Minute'] == updated_time]

    # If there are matching rows, save them to a new CSV file
    if not matched_rows.empty:
        file_name = f"NewFiles2/Acc_Test_{updated_time}.csv"
        matched_rows = matched_rows.drop('Timestamp_Just_Minute', axis = 1)
        matched_rows.to_csv(file_name, index=False)
```
